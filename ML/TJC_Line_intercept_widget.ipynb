```python
import numpy as np
import numpy.linalg as npl
import pandas as pd

import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python

plt.rcParams["figure.figsize"] = (10,10)
vals = np.linspace(-10,10,10000)
```

```python


def callback(a,b,c,a2,b2,c2):
    eq0 = lambda x1: (int(c)-int(a)*x1)/int(b) # 4x1+4x2 = 5
    y0 = eq0(vals)

    eq1 = lambda x1: (int(c2)-int(a2)*x1)/int(b2) # 2x1-4x2 = 1
    y1 = eq1(vals)
    #Intercept ID 
    idx=np.argwhere(np.diff(np.sign(y0 - y1))).flatten()

    try:
      plt.plot(vals,y0,c="blue",label=f"{a}x+{b}y = {c}")
      plt.plot(vals,y1,c="red",label=f"{a2}x+{b2}y = {c2}")  
      plt.plot(vals[idx],y0[idx],'ko',label=f"x={np.round(vals[idx],3)[0]} y={np.round(y0[idx],3)[0]}")
      plt.legend()
    except:
        pass
    plt.show()
    


interact(callback, a="2",b="4",c="5",a2="2",b2="-5",c2="1")
```

```python
eqn="x^2+2y^3=1"
def find_all(string,val):
 return [i for i in range(len(string)) if string.startswith(val, i)]

find_all(eqn,"x")
```

```python

def SplitEqn(eqn):
  #print(eqn)
  a,b,c,n1,n2=1,1,1,1,1
  eqn=eqn.split("=")
  c=int(eqn[~0])
  rhc=eqn[0]
  #Power 
  if "^" in rhc:
      power_i=find_all(rhc,"^")
      n1=rhc[power_i[0]+1]
      if n1=="-":
        n1=-float(rhc[power_i[0]+2])
      else:
        n1=float(n1)
      if len(power_i)==2:
        n2=rhc[power_i[1]]
        if n2=="-":
          n2=- float(rhc[power_i[1]+2])
      else:
        n2=float(n2)
  #Constants 
  #a
  x_i=find_all(rhc.lower(),"x")
 # print(x_i)
  if x_i[0]==1:
    a=float(rhc[x_i[0]-1])
  elif x_i[0]==2: #negative power 
    a=float(rhc[x_i-1])
    if rhc[x_i[0]-2]=="-": #neative power 
      a=-float(a)
  #b
  if len(x_i)==2:
    b=rhc[x_i[1]-1]
    if rhc[x_i[1]-2]=="-": #neative power 
      b=-b
    if b=="+":
      b=float(1)
      
  #print(a,b,c,n1,n2)
  return float(a),float(b),float(c),float(n1),float(n2)

    

def callback(eq0,eq1):

    a,b,c,n1,n2=SplitEqn(eq0)
    EQ0 = lambda x1:   ((c-a*x1**n1)/b)**(1/n2)  #(int(c)-int(a)*x1)/int(b) # 4x1+4x2 = 5
    y0 = EQ0(vals)
    
    a,b,c,n1,n2=SplitEqn(eq1)
    EQ1 = lambda x1: ((c-a*x1**n1)/b)**(1/n2)  # 2x1-4x2 = 1
    y1 = EQ1(vals)
    #Intercept ID 
    idx=np.argwhere(np.diff(np.sign(y0 - y1))).flatten()
    
    #Intrecets
    Inter=[]
    for id in idx:
      Inter.append( [np.round(vals[id],3) , np.round(y0[id],3)]  )

    try:
      plt.plot(vals,y0,c="blue")
      plt.plot(vals,y1,c="red")   
      plt.plot(vals[idx],y0[idx],'ko')
      plt.legend([eq0,eq1,Inter],loc=1)
    except:
        pass
    plt.show()
    
interact(callback, eq0="4x+x=1",eq1="2x^3+3x=0")
```
